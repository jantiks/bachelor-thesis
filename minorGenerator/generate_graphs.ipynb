```python
n = 14
H = {"0a": {"1a", "5a", "3a"},
     "1a": {"0a", "2a", "5a"},
     "2a": {"1a", "3a"},
     "3a": {"0a", "2a", "4a"},
     "4a": {"3a", "5a"},
     "5a": {"0a", "1a", "4a"}}

# Create a mapping from node to index

# Initialize the adjacency matrix

G = {}

for i in range(int(n / len(H))):
    suffix = "b" if i == 1 else "a"
    for key in H:
        new_key = key[:-1] + suffix
        G[new_key] = set()


all_g = []
comb_size = n % len(H)

# Get all combinations
all_combs = list(combinations(H.keys(), comb_size))
for comb in all_combs:
    new_g = G.copy()
    new_g[comb[0][:-1] + "c"] = set()
    new_g[comb[1][:-1] + "c"] = set()    
    all_g.append(new_g)
# Display the adjacency matrix
print(len(all_g), all_combs[0])
```

```python
def color(vertex):
    return vertex[:-1]
def label(vertex):
    return vertex[-1]
```

```python
def exists_kempe(G, start, end, original_begining, looked_at=set()):
    print("ASD START", start, G)
    for key in G[start]:
        if key == end:
            return True
        if color(key) == color(start):
            continue
        if key in looked_at:
            continue
        if ((color(start) != color(end) and color(key) == color(end)) or (color(start) == color(end) and color(key) == color(original_begining))) and exists_kempe(G, key, end, original_begining, looked_at.union({key})):
            return True
    return False
```
